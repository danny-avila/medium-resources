```python
import pandas as pd
import numpy as np
```

```python
prem_league = pd.read_csv('D:/PL Dataset/premierLeague_tables_1992-2017.csv')
```

```python
prem_league.head()
```

```python
prem_league = prem_league[['season', 'team', 'points']]
```

```python
prem_league.head()
```

```python
prem_league.dtypes
```

```python
df = prem_league.pivot_table(values = 'points',index = ['season'], columns = 'team')
```

```python
df.head()
```

```python
df.fillna(0, inplace=True)
df.sort_values(list(df.columns),inplace=True)
df = df.sort_index()
```

```python
df.head()
```

```python
df.iloc[:, 0:-1] = df.iloc[:, 0:-1].cumsum()
```

```python
df[0:10]
```

```python
top_prem_clubs = set()

for index, row in df.iterrows():
    top_prem_clubs |= set(row[row > 0].sort_values(ascending=False).head(6).index)

df = df[top_prem_clubs]
```
